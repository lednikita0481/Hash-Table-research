```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
names = ["csv/1.Return One.csv", "csv/2.ASCII hash.csv", 
         "csv/3.strlen hash.csv", "csv/4.ASCII sum hash.csv",
         "csv/5.ROL hash.csv", "csv/6.ROR hash.csv", "csv/7.Polynom hash.csv"]

colors = ["blue", "blue", "blue", "green", "orange", "grey", "red"]
```

```python
plt.figure(figsize=(30,40), dpi=100)

nrows = 4
ncols = 2

plt.rcParams.update({'font.size': 22})
for i in range(len(names)):
    df = pd.read_csv(names[i], sep = ';')
    index = np.array(df["index"])
    lens = np.array(df["list_len"])
    plt.subplot(nrows, ncols, i+1)
    ax = plt.gca()
    #ax.set_xlim([0, 50])
    plt.bar(index, lens, color = colors[i])
    plt.title(names[i][6:len(names[i]) - 4])
    plt.ylabel("collisions")
    plt.xlabel("hash list")

    print(i)

    if (i > 2):
        plt.subplot(nrows, ncols, ncols*nrows)
        plt.bar(index, lens, color = colors[i])
        plt.title("All approximately normal")
        plt.ylabel("collisions")
        plt.xlabel("hash list")

plt.savefig("graphs.png")
```
